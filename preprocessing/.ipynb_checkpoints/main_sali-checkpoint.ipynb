```python
# the length of each trajectory is 10000
X = data.xtrain
X = X.T
X.shape
```

```python
X0 = X[10]
X0.shape
```

```python
plt.plot(X0)
```

```python
traj_length = 10000
window_size = 1000 # The window length.
n_col = traj_length - window_size + 1 # The number of columns in the trajectory matrix.
n_col
```

```python
# convert series into image

# Create the trajectory matrix by pulling the relevant subseries of F, and stacking them as columns.
Ximg = np.column_stack([X0[i:i+window_size] for i in range(0,n_col)])
```

```python
ax = plt.matshow(Ximg)
plt.xlabel("time-steps")
plt.ylabel("sequence")
plt.colorbar(ax.colorbar, fraction=0.025)
ax.colorbar.set_label("$F(t)$")
plt.title("Hankel Matrix for one trajectory");
```

```python
tmax = np.max(Ximg)
```

```python
import cv2
cv2.saliency
```

```python
saliency = cv2.saliency.StaticSaliencySpectralResidual_create()
```

```python
(success, saliencyMap) = saliency.computeSaliency(Ximg)
```

```python
saliencyMap = (saliencyMap * tmax).astype("uint8")
```

```python
threshMap = cv2.threshold(saliencyMap, 0, tmax, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
```

```python
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1, 3, 1)

ax.set_title('Trajectory ')

plt.axis('off')

plt.imshow(Ximg)


ax=fig.add_subplot(1, 3, 2)

ax.set_title('Saliency Map')

plt.axis('off')

plt.imshow(saliencyMap)


ax=fig.add_subplot(1, 3, 3)

ax.set_title('Anomalies')

plt.axis('off')

plt.imshow(threshMap,cmap="gray")
```
